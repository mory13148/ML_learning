Exercice complet 1 — Iris Dataset (NumPy)

Étape 1 — Télécharger le dataset

```python
import numpy as np

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

data = np.genfromtxt(url, delimiter=',', dtype=str)
data = data[~np.all(data == '', axis=1)]
```

Étape 2 — Explorer le dataset

```python
print("Shape du dataset :", data.shape)
print("5 premières lignes :\n", data[:5])
```

Étape 3 — Séparer features et labels

```python
X = data[:,:-1].astype(float)
y_str  = data[:, -1]
classes = np.unique(y_str)
y = np.array([np.where(classes == label)[0][0] for label in y_str])
print("X_shape : ", X.shape)
print("y shape : ", y.shape)
```

Étape 4 — Normaliser les features

```python
mean = X.mean(axis=0)
std = X.std(axis=0)
X_norm = (X - mean) / std
print(X_norm)
```

Étape 5 — Créer mini-batchs

Essentiel pour utiliser moins de memoire

```python
batch_size = 16
for i in range(0, X_norm.shape[0], batch_size):
    X_batch = X_norm[i:i+batch_size]
    y_batch = y[i:i+batch_size]
    print("Batch X shape :", X_batch.shape, "Batch y shape :", y_batch.shape)
```
